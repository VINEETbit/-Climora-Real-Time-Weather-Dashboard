```python
print("\n--- Outlier Detection ---")

# Identify outliers in 'Quantity' and 'UnitPrice' using IQR method
def find_iqr_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

# Find and display outliers for 'Quantity'
quantity_outliers, q_lower, q_upper = find_iqr_outliers(df, 'Quantity')
print(f"\nOutliers in 'Quantity' (outside {q_lower} to {q_upper}):")
if not quantity_outliers.empty:
    display(quantity_outliers.head())
    print(f"Total Quantity outliers: {len(quantity_outliers)}")
else:
    print("No outliers found for Quantity using IQR method.")

# Find and display outliers for 'UnitPrice'
unitprice_outliers, up_lower, up_upper = find_iqr_outliers(df, 'UnitPrice')
print(f"\nOutliers in 'UnitPrice' (outside {up_lower} to {up_upper}):")
if not unitprice_outliers.empty:
    display(unitprice_outliers.head())
    print(f"Total UnitPrice outliers: {len(unitprice_outliers)}")
else:
    print("No outliers found for UnitPrice using IQR method.")


# Visualize outliers using box plots
print("\n--- Box Plots for Outlier Visualization ---")

plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
sns.boxplot(y=df['Quantity'])
plt.title('Box Plot of Quantity')
plt.ylabel('Quantity')

plt.subplot(1, 2, 2)
sns.boxplot(y=df['UnitPrice'])
plt.title('Box Plot of UnitPrice')
plt.ylabel('UnitPrice')

plt.tight_layout()
plt.show()
```

First, let's clean the data by handling negative values in `Quantity` and `UnitPrice`, as well as missing values in `Description` and `CustomerID`.

```python
# Display the initial shape of the DataFrame
print(f"Initial DataFrame shape: {df.shape}")

# Remove rows with negative Quantity (returns are often handled separately or removed for sales analysis)
df_cleaned = df[df['Quantity'] > 0].copy()
print(f"Shape after removing negative Quantity: {df_cleaned.shape}")

# Remove rows with negative UnitPrice (these are typically errors)
df_cleaned = df_cleaned[df_cleaned['UnitPrice'] > 0].copy()
print(f"Shape after removing negative UnitPrice: {df_cleaned.shape}")

# Handle missing CustomerID (critical for customer-segmentation, so dropping is appropriate)
df_cleaned.dropna(subset=['CustomerID'], inplace=True)
print(f"Shape after dropping rows with missing CustomerID: {df_cleaned.shape}")

# Handle missing Description (dropping them for simplicity, as they are few)
df_cleaned.dropna(subset=['Description'], inplace=True)
print(f"Shape after dropping rows with missing Description: {df_cleaned.shape}")

# Verify data types and non-null counts after cleaning
print("\nDataFrame Info after cleaning:")
df_cleaned.info()

# Display the first few rows of the cleaned DataFrame
print("\nFirst 5 rows of the cleaned DataFrame:")
display(df_cleaned.head())
```

```python
df = df[df["UnitPrice"] > 0]
```

```python
df = df.dropna(subset=["Description"])
```

```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
```

```python
#step 5

df["Revenue"] = df["Quantity"] * df["UnitPrice"]

df.head()

df["Revenue"] = df["Quantity"] * df["UnitPrice"]

df.head()
```

```python
# Calculate 'Revenue' for each transaction
#step 6
df['Revenue'] = df['Quantity'] * df['UnitPrice']

top_products = df.groupby("Description")["Revenue"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))

top_products.plot(kind="bar")

plt.title("Top 10 Products by Revenue")

plt.xlabel("Product")

plt.ylabel("Revenue")

plt.show()
```

```python
#Step 7
country_sales = df.groupby("Country")["Revenue"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))

country_sales.plot(kind="bar")

plt.title("Top Countries by Revenue")

plt.xlabel("Country")

plt.ylabel("Revenue")

plt.show()
```

```python
#STEP 8 – Monthly Sales Trend
#Create month column
df["Month"] = df["InvoiceDate"].dt.to_period("M")
```

```python
#Calculate monthly revenue:
monthly_sales = df.groupby("Month")["Revenue"].sum()
```

```python
#Plot trend:
plt.figure(figsize=(12,5))

monthly_sales.plot()

plt.title("Monthly Revenue Trend")

plt.xlabel("Month")

plt.ylabel("Revenue")

plt.show()
```

```python
#STEP 9 – Quantity vs Revenue Relationship
plt.figure(figsize=(8,5))

plt.scatter(df["Quantity"], df["Revenue"])

plt.title("Quantity vs Revenue Relationship")

plt.xlabel("Quantity")

plt.ylabel("Revenue")

plt.show()
```

```python
# STEP 10 – Customer Distribution by Country
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Group countries: Keep top N, group the rest into 'Other Countries'
customer_dist_raw = df.groupby("Country")["CustomerID"].nunique().sort_values(ascending=False)

top_n_countries = 5 # You can adjust this number
customer_dist_top = customer_dist_raw.head(top_n_countries)
other_sum = customer_dist_raw.iloc[top_n_countries:].sum()

if other_sum > 0:
    customer_dist = pd.concat([customer_dist_top, pd.Series({'Other Countries': other_sum})])
else:
    customer_dist = customer_dist_top

plt.figure(figsize=(12,12)) # Increased figure size for more spacing

# Create explode effect for the largest slice (United Kingdom)
explode_effect = [0.1 if country == 'United Kingdom' else 0 for country in customer_dist.index]

# Define a professional color palette
colors = sns.color_palette('viridis', n_colors=len(customer_dist)) # A visually appealing sequential palette

# Plotting the pie chart with enhanced aesthetics
wedges, texts, autotexts = plt.pie(customer_dist.values,
                                   autopct='%1.1f%%',
                                   startangle=90,
                                   explode=explode_effect,
                                   shadow=True,
                                   pctdistance=0.85, # Distance of percentages from center
                                   colors=colors,
                                   wedgeprops={'edgecolor': 'black', 'linewidth': 0.5}) # Add thin black borders

# Adjust font size and color for the percentage labels (inside the pie)
for autotext in autotexts:
    autotext.set_fontsize(12)
    autotext.set_color('white') # Changed to white for better contrast
    autotext.set_fontweight('bold')

plt.title("Customer Distribution by Country", fontsize=18, fontweight='bold', pad=20)

# Add a clear legend for country names
plt.legend(wedges, customer_dist.index, title="Countries", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), fontsize=12, title_fontsize=14)

plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a perfect circle.
plt.tight_layout() # Adjust layout to prevent labels/legend from being cut off
plt.show()
```
